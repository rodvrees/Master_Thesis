```python
data=data[data["RSA"].notna()]
data=data[~data["RSA"].str.contains("no")]
```

```python
data
```

```python
import re
data["RSA"] = pd.to_numeric(data["RSA"])
pat = re.compile(r"\](.*?)\[")
data["Mod"] = data["modifications"].apply(lambda x: re.search(pat, x)[1].capitalize())
```

```python
data
```

```python
data = data[data["Site_occupancy"] != 1]
```

```python
ax = sns.regplot(data=data, x="Site_occupancy", y="RSA", ci=None, line_kws={"color": "r"})
ax.set_title("All oxPTMs")
```

```python
for mod in data["Mod"].unique():
    moddata = data[data["Mod"] == mod]
    sns.lmplot(data=moddata, x="Site_occupancy", y="RSA", line_kws={"color":"r"}, ci=False)
    ax = plt.gca()
    ax.set_title(mod)
```

```python
data["Evaluation"] = np.where(data.RSA > 0.2, 'Accessible', 'Buried')
```

```python
data
```

```python
relevant = data[data["Mod"].isin(["Oxidation", "Dioxidation", "Trioxidation", "Carbonyl"])]
```

```python
hue_plot_params = {'data':relevant, "x":'Mod', "y":'Site_occupancy', "hue": "Evaluation"}
from statannotations.Annotator import Annotator
pairs = [[('Oxidation', 'Buried'), ('Oxidation', 'Accessible')], [('Dioxidation', 'Buried'), ('Dioxidation', 'Accessible')], [('Carbonyl', 'Buried'), ('Carbonyl', 'Accessible')], [('Trioxidation', 'Buried'), ('Trioxidation', 'Accessible')]]
ax = sns.boxplot(data=relevant, x="Mod", y="Site_occupancy", hue="Evaluation")
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
annotator = Annotator(ax, pairs, **hue_plot_params)
annotator.configure(test="Mann-Whitney", text_format="simple")
annotator.apply_and_annotate()
ax.set_ylabel("Site Occupacy")
ax.set_xlabel("Modification")
plt.show()
```

```python
Oxidationaccessible = relevant[(relevant["Mod"] == "Oxidation") & (relevant["Evaluation"] == "Accessible")]["Site_occupancy"].to_list()
Oxidationburied = relevant[(relevant["Mod"] == "Oxidation") & (relevant["Evaluation"] == "Buried")]["Site_occupancy"].to_list()
Dioxidationaccessible = relevant[(relevant["Mod"] == "Dioxidation") & (relevant["Evaluation"] == "Accessible")]["Site_occupancy"].to_list()
Dioxidationburied = relevant[(relevant["Mod"] == "Dioxidation") & (relevant["Evaluation"] == "Buried")]["Site_occupancy"].to_list()
Carbonylaccessible = relevant[(relevant["Mod"] == "Carbonyl") & (relevant["Evaluation"] == "Accessible")]["Site_occupancy"].to_list()
Carbonylburied = relevant[(relevant["Mod"] == "Carbonyl") & (relevant["Evaluation"] == "Buried")]["Site_occupancy"].to_list()
Trioxidationaccessible = relevant[(relevant["Mod"] == "Trioxidation") & (relevant["Evaluation"] == "Accessible")]["Site_occupancy"].to_list()
Trioxidationburied = relevant[(relevant["Mod"] == "Trioxidation") & (relevant["Evaluation"] == "Buried")]["Site_occupancy"].to_list()
```

```python
from scipy.stats import mannwhitneyu
U, p = mannwhitneyu(Oxidationaccessible, Oxidationburied, nan_policy="omit")
print(p)
```

```python
U, p = mannwhitneyu(Dioxidationaccessible, Dioxidationburied, nan_policy="omit")
print(p)
```

```python
U, p = mannwhitneyu(Carbonylaccessible, Carbonylburied, nan_policy="omit")
print(p)
```

```python
U, p = mannwhitneyu(Trioxidationaccessible, Trioxidationburied, nan_policy="omit")
print(p)
```
