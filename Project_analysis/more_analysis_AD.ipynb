```python
import os
os.chdir("/home/robbe/ionbot")
import pandas as pd
import OxiAnalysis as OA
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
quant = pd.read_csv("/home/robbe/ionbot/ionbot_0.9.5/PXD016278/flashlfq/QuantifiedPeptides.tsv", sep="\t")
quant.drop(list(quant.filter(regex = 'Detection Type')), axis = 1, inplace = True)
quant = quant[~quant["Protein Groups"].str.contains("sp|", na=False, regex=False)]

quant.head()
```

```python
cols = [col for col in quant.columns if "Intensity" in col]
```

```python
quant = OA.quantile_transform(quant, cols)
```

```python
quant.head()
```

```python
quant["modifications"] = quant["Sequence"].apply(OA.flashLFQmods)
quantex = quant.explode("modifications")
quantex.drop(list(quantex.filter(regex = 'Detection Type')), axis = 1, inplace = True)
quantex.drop(["Gene Names", "Organism", "Unnamed: 101"], axis=1, inplace=True)
data = quantex.groupby("modifications").sum()
```

```python
def splitatn(strng, sep, pos):
        strng = strng.split(sep)
        return sep.join(strng[:pos]), sep.join(strng[pos:])
```

```python
data.rename(columns= lambda x: x.split("_")[11] + x.split("_")[12], inplace=True)
```

```python
data = data.loc[~(data==0).all(axis=1)]
```

```python
data.drop(axis=0, labels="None", inplace=True)
```

```python
data.sum(axis=1)
```

```python
totalintensities = data.sum(axis=1).to_frame().reset_index()
```

```python
sorted = totalintensities.sort_values(0, ascending=False).reset_index()
```

```python
sorted["Rank"] = sorted.index+1
```

```python
sorted.rename(mapper={0:"Abundance"}, axis=1, inplace=True)
```

```python
sorted
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.lineplot(data=sorted, x="Rank", y="Abundance", markers=True, marker="s", markevery=[0,1,3,4,5,199,599,898])
plt.ylabel("Modification abundance")
plt.title("Most modification are very low in abundance")
for x,y,m in sorted[["Rank","Abundance","modifications"]].values:
    if x < 6 or x == 200 or x == 600 or x ==899:
        if not m == "semi_tryptic":
            ax.text(x+10,y+5000,m, fontsize=8)
```

```python
total_abundance= sorted["Abundance"].sum()
sorted["Proportion"] = sorted["Abundance"] / total_abundance
```

```python
sorted
```

```python
total_proportion = 0
for i, row in sorted.iterrows():
    total_proportion += row['Proportion']
    if total_proportion >= 0.5:
        print("Number of rows needed for 50% proportion:", i+1)
        break
```
