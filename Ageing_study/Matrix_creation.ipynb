```python
def splitatn(strng, sep, pos):
    strng = strng.split(sep)
    return sep.join(strng[:pos]), sep.join(strng[pos:])
```

```python
relevant_SCG["id"] = relevant_SCG["spectrum_file"].apply(lambda x: splitatn(x,"_",2)[1])
```

```python
young = relevant_SCG[relevant_SCG["spectrum_file"].str.contains("VUR")]
old = relevant_SCG[relevant_SCG["spectrum_file"].str.contains("RRAG")]
```

```python
old
```

```python
young["Subject"] = young["id"].str.split("_").str[1].str[:-4]
old["Subject"] = old["id"].str.split("_").str[0].str[0:3]
```

```python
alldata = pd.concat([young,old])
```

```python
alldata
```

```python
alldata_with_reps = alldata.groupby(["Modification", "Subject"]).median().reset_index()
```

```python
alldata_with_reps.groupby(["Modification", "Subject"]).sum()
```

```python
pivot_df = pd.pivot_table(alldata_with_reps, values="RelCount", index="Modification", columns="Subject").fillna(0)
```

```python
row_mins = pivot_df.min(axis=1)
row_maxs = pivot_df.max(axis=1)

normalized_df = pivot_df.apply(lambda row: (row - row_mins[row.name])/(row_maxs[row.name] - row_mins[row.name]), axis=1)
```
