```python
for path in tsv_files:
    tsv = pd.read_csv(path, sep='\t')
    Output_df2 = pd.merge(Output_df2, tsv,
                          on=['Sequence', 'Base Sequence', 'Protein Groups', 'Gene Names', 'Organism'], how='outer')
```

```python
df = Output_df2[Output_df2.columns.drop(list(Output_df2.filter(regex='Detection|Unnamed')))]
```

```python
cols = [i for i in df.columns if "Intensity" in i]
```

```python
OA.quantile_transform(df, cols)
```

```python
df = df[df.columns.drop(list(df.filter(regex='Gene|Organism')))]
```

```python
df
```

```python
dfnonan = df[df['Protein Groups'].notna()]
```

```python
dfnonan
```

```python
dfnonan["Values"] = dfnonan.count(axis=1)
```

```python
dfnonan50 = dfnonan[dfnonan["Values"] >= 53]
```

```python
dfnonan50
```

```python
nonmodifieds = dfnonan50[dfnonan50["Sequence"].str.endswith('nan')]
```

```python
nonmodifieds
```

```python
modifieds = dfnonan50[~dfnonan["Sequence"].str.endswith('nan')]
```

```python
modifieds
```

```python
modifieds["Modifications"] = modifieds["Sequence"].apply(OA.flashLFQmods)
```

```python
oxmodifieds = modifieds[modifieds["Modifications"].isin(OA.modslist)]
```
