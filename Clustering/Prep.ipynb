```python
import os
os.chdir("/home/robbe/ionbot/")
import OxiAnalysis as OA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PXD022545 = pd.read_csv("/home/robbe/ionbot/ionbot_0.9.5/PXD022545/new_quatn/QuantifiedPeptides.tsv", sep="\t")
PXD012477 = pd.read_csv("/home/robbe/ionbot/ionbot_0.9.5/PXD012477/flashlfq_results/QuantifiedPeptides.tsv", sep="\t")
PXD014381 = pd.read_csv("/home/robbe/ionbot/ionbot_0.9.5/PXD014381/mzml_files/QuantifiedPeptides.tsv", sep="\t")
PXD010469 = pd.read_csv("/home/robbe/ionbot/ionbot_0.9.5/PXD010469/quant/QuantifiedPeptides.tsv", sep="\t")
PXD004713 = pd.read_csv("/home/robbe/ionbot/ionbot_0.9.5/PXD004713/quant/QuantifiedPeptides.tsv", sep="\t")
syn21441775 = pd.read_csv("/home/robbe/ionbot/ionbot_0.9.5/Synapse_Aging/quant/QuantifiedPeptides.tsv", sep="\t")
datalist = [PXD022545, PXD012477, PXD014381, PXD004713, PXD010469, syn21441775]
```

```python
data = {}
for index, dataset in enumerate(datalist):
    cols = [col for col in dataset.columns if "Intensity" in col]
    OA.quantile_transform(dataset, cols)
    dataset[cols] = dataset[cols].transform(lambda x: x/x.sum())
    dataset.drop(list(dataset.filter(regex = 'Detection Type|Unnamed|gis')), axis = 1, inplace = True)
    SI = OA.summedintensities(dataset)
    SI = SI.drop(columns=["Organism", "Gene Names", "Oxmod?"])
    data[index] = SI
```

```python
def merge(frames, on_columns):
    #implementation here
    if not frames:
        return None
    if len(frames) == 1:
        return frames[0]
    out = frames[0]
    for df in frames[1:]:
        out = out.merge(df, on=on_columns, how="outer")
    return out
```

```python
frames = [data[i] for i in range(len(data))]
```

```python
mergeddata = merge(frames, on_columns="Modifications")
```

```python
mergeddata = mergeddata.set_index("Modifications")
```

```python
mergeddataT = mergeddata.T
```
